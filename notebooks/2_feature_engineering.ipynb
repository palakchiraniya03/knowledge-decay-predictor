```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/study_memory_data.csv")
df
```

```python
df["revision_density"] = df["revision_count"] / (df["days_since_last_revision"] + 1)

df["difficulty_impact"] = df["difficulty"] * (1 - df["memory_strength"])

df["performance_gap"] = 100 - df["last_score"]

df
```

```python
features = [
    "days_since_last_revision",
    "revision_count",
    "difficulty",
    "last_score",
    "revision_density",
    "difficulty_impact",
    "performance_gap"
]

X = df[features]
y = df["memory_strength"]
X, y
```

```python
processed_df = df.copy()
processed_df.to_csv("../data/processed_study_memory_data.csv", index=False)
```
