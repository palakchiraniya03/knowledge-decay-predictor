```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from src.decay_recommender import get_recommendation
```

```python
df = pd.read_csv("../data/processed_study_memory_data.csv")
df
```

```python
features = [
    "days_since_last_revision",
    "revision_count",
    "difficulty",
    "last_score",
    "revision_density",
    "difficulty_impact",
    "performance_gap"
]

X = df[features]
y = df["memory_strength"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
df["predicted_memory_strength"] = model.predict(X)
df
```

```python
df["recommendation"] = df["predicted_memory_strength"].apply(get_recommendation)
df
```

```python
final_output = df[["topic", "predicted_memory_strength", "recommendation"]]
final_output
```
